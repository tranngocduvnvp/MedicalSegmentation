```python
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
root = "/home/bigdata/Documents/TND_Modeling/dataset"
DATA_NAME = ["Montgomery", "Shenzhen"]
list_img_paths = [glob.glob(f"{root}/{name}/img/*") for name in DATA_NAME]
```

```python
labels_monto = np.array([int(file.split("_")[-1].split(".")[0]) for file in list_img_paths[0]])
labels_shen = np.array([int(file.split("_")[-1].split(".")[0]) for file in list_img_paths[1]])
```

```python
values, count = np.unique(labels_monto, return_counts=True)
print(count)
plt.bar(["normal", "abnormal"], count)
plt.xticks([0,1])
plt.title("Montgomery dataset")
plt.bar_label(plt.gca().containers[0])
```

```python
values, count = np.unique(labels_shen, return_counts=True)
print(count)
plt.bar(["normal", "abnormal"], count)
plt.xticks([0,1])
plt.title("Shenzhen dataset")
plt.bar_label(plt.gca().containers[0])
```
